```python
import subprocess
import pandas as pd
import matplotlib.pyplot as plt
```

# Insertion sort

```python
resultado_insertion = subprocess.run(["ordenacao\insertion\insertion.exe"], capture_output=True, text=True)
print(resultado_insertion.stdout[:300])
```

```python
linhas = resultado_insertion.stdout.strip().split("\n")

valores = []
for linha in linhas:
    tipo, n, tempo = linha.split()
    valores.append((tipo, int(n), float(tempo)))

df = pd.DataFrame(valores, columns=["tipo", "tamanho", "tempo_ms"])
```

```python
for tipo in df["tipo"].unique():
    subset = df[df["tipo"] == tipo]
    plt.plot(subset["tamanho"], subset["tempo_ms"], marker="o", label=tipo)
    
plt.xlabel("Tamanho da entrada(n)")
plt.ylabel("Tempo de execuo (ms)")
plt.title("Tempo de execuo do Insertion Sort")
plt.grid(True)
plt.legend()
plt.show()
```
